```python
import mikeio
import pandas as pd
import os
import glob
from mikeio.eum import EUMType, EUMUnit, ItemInfo

# Output folder path (Just the folder path)
output_folder_path = r'G:\MJI\BWDB_Tidal (104 nos.)\Output\Fortnightly_Avg'


for file in glob.glob(r'G:\MJI\BWDB_Tidal (104 nos.)\Output\Fortnightly_Avg\*.xlsx'):

    # Splitting the path by seperator
    path_list = file.split(os.sep)  

    # File absolute path with file name and extension as xlsx
    df = pd.read_excel(file)
    df['Date/Time'] = pd.to_datetime(df['Date/Time'], format = "%Y/%m/%d")

    # Setting 'Date' column as dataframe index
    df.set_index(df['Date/Time'], inplace=True)
    v_df = df['Modified_Cyclone']

    # Output file path and name with extension
    file_path = os.path.join(output_folder_path, '{}.dfs0'.format(path_list[-1][:-5]))

    # Defining dfs0 item with item type
    item = [ItemInfo('Modified_Cyclone', EUMType.Water_Level, EUMUnit.meter)]

    # Saving the dfs0 file
    v_df.to_dfs0(file_path,items=item,title='{}'.format(path_list[-1][:-21]))
    
    print(path_list[-1][:-5])
```
