```python
import mikeio
import pandas as pd
import numpy as np
import math
import os
import glob

# Reading the template pfs (paramter file system) file
pfs = mikeio.read_pfs(r"G:\MJI\BWDB_Tidal (104 nos.)\Notebooks\SW5 (Madaripur).21t")

# Reading the station co-ordinate excel file
df = pd.read_excel(r"G:\MJI\BWDB_Tidal (104 nos.)\Data Quality Inventory of BWDB WL Station Data_Tidal.xlsx")

for file in glob.glob(r'G:\MJI\BWDB_Tidal (104 nos.)\Output\Tidal_Constituents\*dfs0'):
    
    path_list = file.split(os.sep)
    
    # Reading the equidistant timeseries dfs0
    ds = mikeio.read(file)

    # Coverting the datset file to pandas dataframe
    ts_df = ds.to_dataframe()
    
    # Finding the total number of timestep in the dataset
    total_time_step = len(ts_df)
    
    # Changing the equidistant timeseries dfs0 file name in the parameter file
    pfs.m21hd_tideanalysisheights.Setup.InputFile.DataFile.FileName = '|.\\{}.dfs0|'.format(path_list[-1][:-5])
    
    # Changing the total number of time step in the pfs file
    pfs.m21hd_tideanalysisheights.Setup.SubSeries.SUB_SERIES_LAST = total_time_step - 2
    
    # Setting the lattitude location in the pfs file
    station_name = path_list[-1][:-14]
    lat_array = (df[df['Station ID'] == station_name]['Latitude']).values
    pfs.m21hd_tideanalysisheights.Setup.GeneralParameters.Latitude = lat_array[0]
    
    # Setting the constitute file name
    pfs.m21hd_tideanalysisheights.Setup.Output.ConstituentsFileName = '|.\\{}.con|'.format(path_list[-1][:-14])
    
    # writing the pfs file and saving it
    pfs.write(r'G:\MJI\BWDB_Tidal (104 nos.)\Output\Tidal_Constituents\{}.21t'.format(path_list[-1][:-14]))
    
    print(path_list[-1][:-14])
```
