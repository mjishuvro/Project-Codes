```python
import pandas as pd
import numpy as np
import os
import glob

tmax_df = pd.read_csv(r"G:\MJI\SKZ\Script\From RAJ\TmaxData.csv")
tmin_df = pd.read_csv(r"G:\MJI\SKZ\Script\From RAJ\TminData.csv")
grid_df = pd.read_excel(r"G:\MJI\SKZ\Script\From RAJ\CMIP6_grid_dharla_river.xlsx")


scenario_list = ['historical', 'ssp245', 'ssp585']
model_list = ['ACCESS-CM2', 'ACCESS-ESM1-5', 'BCC-CSM2-MR', 'CanESM5', 'EC-Earth3', 'EC-Earth3-Veg', 
              'INM-CM4-8', 'INM-CM5-0', 'MPI-ESM1-2-HR', 'MPI-ESM1-2-LR', 'MRI-ESM2-0', 'NorESM2-LM', 'NorESM2-MM']

points_list = [f"('{lon}', '{lat}')" for lon, lat in zip(grid_df['Longitude'], grid_df['Latitude'])]

# Iterating through each scenario and creating scenario-specific dataframe
for scenario in scenario_list:
    try:
        os.makedirs(f'Output/{scenario}')

    except:
        pass
    
    tmax_scenario_df = tmax_df[tmax_df['projection'] == scenario]
    tmin_scenario_df = tmin_df[tmin_df['projection'] == scenario]

    
    # Iterating through each model and creating model-specific dataframe
    for model in model_list:
        try:
            os.makedirs(f'Output/{scenario}/{model}')

        except:
            pass
        
        tmax_model_df = tmax_scenario_df[tmax_scenario_df['model'] == model]
        tmin_model_df = tmin_scenario_df[tmin_scenario_df['model'] == model]


        # Defining the grid points file and generating the input file
        tmp_file = open(f'Output/{scenario}/{model}/tmp.txt', 'a')
        tmp_file.write('ID,NAME,LAT,LONG,ELEVATION' + '\n')

        
        # Iterating through each grid point and creating input file from each point
        for i in range(0, len(points_list)):

            file_name = open(f'Output/{scenario}/{model}/tmp{i+1}.txt', 'a')

            point_tmax_df = tmax_model_df.loc[:, ['date', points_list[i]]]
            point_tmin_df = tmin_model_df.loc[:, ['date', points_list[i]]]
            point_temperature_df = pd.merge(point_tmax_df, point_tmin_df, on='date', suffixes=('_tmax', '_tmin'))
        
            first_date = (point_temperature_df.iloc[0,0]).replace('-', '')   ## yyyymmdd
            file_name.write(first_date + '\n')
            
            for r in range(0, len(point_temperature_df)):
                max_temp = point_temperature_df.iloc[r,1]
                min_temp = point_temperature_df.iloc[r,2]
                file_name.write(f'{max_temp},{min_temp}\n')

            # Updating the point location the grid file
            LAT = float(eval(points_list[i])[1])
            LONG = float(eval(points_list[i])[0])
            NAME = f'tmp{i+1}'
            
            tmp_file.write(f'{i+1},{NAME},{LAT},{LONG},0.00\n')
```
