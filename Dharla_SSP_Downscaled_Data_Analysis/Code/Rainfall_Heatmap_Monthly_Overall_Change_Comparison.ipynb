```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import seaborn as sns
import datetime
import calendar

df = pd.read_csv(r"D:\Script\From_RAJ\Code\PrecipData.csv")

df['date'] = pd.to_datetime(df['date'])
df['Year'] = df['date'].dt.year
df['Month'] = df['date'].dt.month
df['Day'] = df['date'].dt.day

timespan_dict = {'Base': [1990, 2014],
                 'Short': [2026, 2050],
                 'Mid': [2051, 2075],
                 'Long': [2076, 2100]}

ts_conditions = [
                (df['date'].dt.year >= timespan_dict['Base'][0]) & (df['date'].dt.year <= timespan_dict['Base'][1]),
                (df['date'].dt.year >= timespan_dict['Short'][0]) & (df['date'].dt.year <= timespan_dict['Short'][1]),
                (df['date'].dt.year >= timespan_dict['Mid'][0]) & (df['date'].dt.year <= timespan_dict['Mid'][1]),
                (df['date'].dt.year >= timespan_dict['Long'][0]) & (df['date'].dt.year <= timespan_dict['Long'][1])
                ]

ts_choices = list(timespan_dict.keys())

df['time_span'] = np.select(ts_conditions, ts_choices, default='Not_Included')
df['Spatial_Mean']= df[df.columns[1:-8]].mean(axis=1)
spatial_mean_df = df[['time_span', 'model', 'projection', 'Year', 'Month', 'Day', 'date', 'Spatial_Mean']]

monthly_sum_df = spatial_mean_df.groupby(['time_span', 'model', 'projection', 'Year', 'Month'])['Spatial_Mean'].sum().reset_index()
monthly_sum_df.rename({'Spatial_Mean' : 'Monthly_Sum'}, inplace = True, axis=1)
selected_monthly_df = monthly_sum_df[monthly_sum_df['time_span']!= 'Not_Included']

monthly_mean_df = selected_monthly_df.groupby(['time_span', 'projection', 'Month'])['Monthly_Sum'].mean().reset_index()
model_base_df = monthly_mean_df[monthly_mean_df['projection'] == 'historical']
merged_monthly_mean_df = pd.merge(monthly_mean_df, model_base_df[['Month', 'Monthly_Sum']], on=['Month'], suffixes=('', '_base'))
change = ((merged_monthly_mean_df['Monthly_Sum'] - merged_monthly_mean_df['Monthly_Sum_base'])*100)/merged_monthly_mean_df['Monthly_Sum_base']
merged_monthly_mean_df['Change (%)'] = change
merged_monthly_mean_df = merged_monthly_mean_df[merged_monthly_mean_df['Change (%)'] != 0]

# Colormap for precipitation changes
cmap_precip = plt.get_cmap("Blues")  

for scenario in set(merged_monthly_mean_df['projection']):
    scenario_df = merged_monthly_mean_df[merged_monthly_mean_df['projection'] == scenario]
    selected_scenario_df = scenario_df[['time_span', 'Month', 'Change (%)']]

    heatmap_df = selected_scenario_df.pivot(index='time_span', columns='Month', values='Change (%)')
    heatmap_df.sort_index(ascending=False, inplace=True)

    fig, ax = plt.subplots(1,1,figsize=(8,4), dpi=300)
    font = {'family' : 'Times New Roman',              
            'weight' : 'normal',
            'size'   : 10}
    
    plt.rc('font',**font)
    plt.rc('xtick',labelsize=10)
    plt.rc('ytick',labelsize=10)

    sns.heatmap(data=heatmap_df, vmin=None, vmax=None, annot=True, fmt='.1f',  cmap=cmap_precip, square=True,
                cbar=False, linewidths=0.1, linecolor='black', ax=ax, annot_kws={'color': 'black'})
    
    # Customize x-tick labels (months)
    xtick_positions = ax.get_xticks()
    xtick_labels_list = [calendar.month_name[int(label.get_text())][:3] for label in ax.get_xticklabels()]
    ax.set_xticks(xtick_positions)
    ax.set_xticklabels(xtick_labels_list)
    
    # Customize the y-tick labels (timespan)
    timespan_labels = ['Near\n(2026-2050)', 'Mid\n(2051-2075)', 'Long\n(2076-2100)']
    ax.set_yticks(np.arange(len(timespan_labels)) + 0.5) 
    ax.set_yticklabels(timespan_labels, fontsize=10, rotation=0, ha='center', va='center')
    ax.tick_params(axis='y', which='major', pad=27)
    
    # Move x-axis to top
    ax.xaxis.tick_top()
    ax.xaxis.set_label_position('top')
    ax.yaxis.set_label_position('left')
    
    # Hide tick lines
    ax.tick_params(axis='x', which='both', length=0)
    ax.tick_params(axis='y', which='both', length=0)
    
    # Show ticks on top and left
    ax.tick_params(top=True, bottom=False, left=True, right=False)
    ax.xaxis.set_ticks_position('top')
    ax.yaxis.set_ticks_position('left')
    
    # Adding the colorbar
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("bottom", size="10%", pad=0.1)  # 5% height, 0.3 pad
    sm_precip = plt.cm.ScalarMappable(cmap=cmap_precip, norm=plt.Normalize(vmin=-20, vmax=60))
    cbar = plt.colorbar(sm_precip, cax=cax, orientation='horizontal', extend='both')
    cbar.set_label('Change in Monthly Precipitation (%)', fontsize=10)
    
    ax.xaxis.set_label_position('top') 
    ax.set_xlabel('Month', fontsize=12, fontweight='bold')
    ax.set_ylabel('Timespan',  fontsize=12, fontweight='bold')
    ax.set_title(f'Monthly Change in Rainfall ({scenario.upper()})', fontsize=12, 
                 fontweight='bold', x=0.5, y = 1.25)
    
    plt.tight_layout()
    plt.savefig(r'D:\Script\From_RAJ\Output\Precipitation_{}_Heatmap.jpg'.format(scenario),
                dpi=600, bbox_inches = 'tight')    
```
