```python
import mikeio
import pandas as pd
import numpy as np
import os
import glob


def item_wise_column_list(dataframe):
    
    df = dataframe

    all_item = []

    for x in list(df.columns)[1:]:
        y = x.split(":")
        all_item.append(y[-1])

    unique_list = list(set(all_item))    


    datalists = dict()
    for item in unique_list:
        name = '{}'.format(item)
        datalists[name] = []


    for col in list(df.columns)[1:]:

        item_name = c_name = col.split(':')[-1]

        if item_name == [name for name in datalists.keys() if name == item_name][0]:

            datalists[item_name].append(col)

        else:
            pass

    return datalists




l_df = pd.read_excel(r"G:\MJI\SKZ\Script\Storm Surge Data Extrcation from DFS0 File\Teknaf_Data_Extraction\Storm_Surge\Location_ID.xlsx")

max_item_name = ' Surface elevation'

Location_list = []

for location in l_df['ID']:
    Location_list.append(str(location))


for place in Location_list:

    r_df = pd.DataFrame()


    for file in glob.glob(r"G:\MJI\SKZ\Script\Storm Surge Data Extrcation from DFS0 File\Teknaf_Data_Extraction\Storm_Surge\CC\*.dfs0"):
        
        ds = mikeio.read(file)

        df = ds.to_dataframe()

        df.reset_index(inplace=True)

        df.rename(columns= {'index':'Date'},inplace = True)


        
        item_dict = item_wise_column_list(df)
        item_name = place
        item_df_col_list = [name for key in item_dict.keys() for name in item_dict[key] if name.split(':')[0] == item_name]
        
        df_col_list = ['Date']
        
        sup_list = [df_col_list.append(i) for i in item_df_col_list]
        
        s_df = df[df_col_list]

        n_df = s_df.iloc[12:,:]
        
        col_max_item_name = '{}:{}'.format(place, max_item_name)

        f_df = n_df.nlargest(1,col_max_item_name)
        
        f_df['Max_Field'] = max_item_name

        r_df = pd.concat([r_df,f_df])

        r_df.to_excel(r'G:\MJI\SKZ\Script\Storm Surge Data Extrcation from DFS0 File\Teknaf_Data_Extraction\Storm_Surge\CC\Teknaf_CC_{}.xlsx'.format(place))
```
