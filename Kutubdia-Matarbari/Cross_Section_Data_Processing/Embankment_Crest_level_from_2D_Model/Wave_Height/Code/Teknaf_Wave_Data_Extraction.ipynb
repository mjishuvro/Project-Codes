```python
import mikeio
import os
import glob
import pandas as pd
import numpy as np


file_list = []
time_step_list = []

df = pd.read_excel(r"G:\MJI\SKZ\Script\Storm Surge Data Extrcation from DFS0 File\Teknaf_Data_Extraction\Wave\Wave_Result_File_List_Modified.xlsx")

for name,time in zip(df['File_Name'],df['Time_Step']):
    file_name = '{}.dfsu'.format(name)
    file_list.append(file_name)
    time_step_list.append(time)
    
file_df = pd.DataFrame({'File_Name':file_list,'Time_Step':time_step_list})




pfs = mikeio.read_pfs(r"G:\MJI\SKZ\Script\Storm Surge Data Extrcation from DFS0 File\Teknaf_Data_Extraction\Wave\Teknaf_E_742922_N_296337.dxfm")


out_folder = r'G:\MJI\SKZ\Script\Storm Surge Data Extrcation from DFS0 File\Teknaf_Data_Extraction\Wave\Output'

for file in glob.glob(r'\\10.0.0.33\New folder (D)\P51624_long_term_Wave\Wave Model\SIM\**\**\*.dfsu'):
    

    folder_path = file.split(os.sep)
    dfsu_file_name = folder_path[-1]
    name1 = folder_path[-1].split('.dfsu')
    name = name1[0]
    

    if dfsu_file_name in file_list:
        
        n_df = file_df[file_df['File_Name'] == dfsu_file_name]
        time = int(n_df['Time_Step'].values)
        
        input_folder = r'\\10.0.0.33\New folder (D)\P51624_long_term_Wave\Wave Model\SIM\{}\{}\{}.dfsu'.format(folder_path[-3],folder_path[-2],name)


        output_all_key_dict = pfs.DataExtractionFM.OUTPUTS.keys()
        output_list = list(output_all_key_dict)[3:]

        for output_x in output_list:
            
            pfs.DataExtractionFM.OUTPUTS[output_x].last_time_step = time-1
            
            x = pfs.DataExtractionFM.OUTPUTS[output_x].file_name.split('_')[0]
            y = pfs.DataExtractionFM.OUTPUTS[output_x].file_name.split('_')[-1][-6:]

            r_output = '{}_{}_{}'.format(x,name,y)

            pfs.DataExtractionFM.OUTPUTS[output_x].file_name = r_output


        dxfm_file_input = '|{}|'.format(input_folder)

        pfs.DataExtractionFM.INPUT.file_name = dxfm_file_input

        dxfm_out_folder = os.path.join(out_folder,name)

        os.mkdir(dxfm_out_folder)

        dxfm_file_name = r'{}\{}.dxfm'.format(dxfm_out_folder, name)

        pfs.write(dxfm_file_name)

        print(name)

    else: 
        pass
```
