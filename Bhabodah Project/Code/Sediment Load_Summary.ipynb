```python
import pandas as pd
import numpy as np
import datetime
import glob
import os
```

```python
sediment_load_summary_df = pd.DataFrame()

for file in glob.glob(r'..\..\Output (Dry Season)\Sediment_Load\*.xlsx'):

    file_path_list = file.split(os.sep)
    river_name = file_path_list[-1].split('_')[0]
    station_name = file_path_list[-1].split('_')[1]
    tidal_phase = file_path_list[-1].split('_')[2]
    
    df = pd.read_excel(file)

    station_sediment_summary_df = df.groupby('Flow Type')['Sediment Load (million kg)'].sum().to_frame().T
    station_sediment_summary_df.rename(columns={'Ebb': 'Ebb (million kg)', 'Flood': 'Flood (million kg)'}, inplace=True)
    station_sediment_summary_df.reset_index(drop=True, inplace=True)

    station_sediment_summary_df.insert(loc=0, column = 'River', value = river_name)
    station_sediment_summary_df.insert(loc=1, column = 'Station Name', value = station_name)
    station_sediment_summary_df.insert(loc=2, column = 'Tidal Phase', value = tidal_phase)
    station_sediment_summary_df.insert(loc=3, column = 'Date', value = df['Date_Time'].dt.date.unique())

    sediment_load_summary_df = pd.concat([sediment_load_summary_df, station_sediment_summary_df])

sediment_load_summary_df['Net Sediment Load (million kg)'] = sediment_load_summary_df['Ebb (million kg)'].fillna(0) + sediment_load_summary_df['Flood (million kg)'].fillna(0)
sediment_load_summary_df['Net Sediment Load (tons)'] = sediment_load_summary_df['Net Sediment Load (million kg)']*1000
sediment_load_summary_df.to_csv(r'..\..\Output (Dry Season)\Dry_Sediment_Load_Summary.csv', float_format='%.4f', index=False)
sediment_load_summary_df
```
